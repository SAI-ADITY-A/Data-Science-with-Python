#Importing the Dependencies and creating a dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout
```

```python
x_train = np.linspace(-1, 1, 20)
```

```python
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  ,
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475,
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
x_test = np.linspace(-1, 1, 20)
```

```python
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(x_train, y_train, color = 'red', label = 'Train')
plt.scatter(x_test, y_test, color = 'black', label = 'Test')
plt.legend()
plt.show()
```

#Regression model

```python
ann = Sequential()
```

```python
ann.add(Dense(units = 128, activation = 'relu', input_dim = 1))
ann.add(Dense(units = 128, activation = 'relu'))
ann.add(Dense(units = 1, activation = 'linear'))
```

```python
ann.compile(optimizer = 'adam', loss = 'mse', metrics = ['accuracy'])
```

```python
history = ann.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 500)
```

```python
y_pred = ann.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color = 'red', label = 'Train')
plt.scatter(x_test, y_test, color = 'black', label = 'Test')
plt.plot(x_test, y_pred, color = 'blue', label = 'Prediction')
plt.legend()
plt.show()
```

#Dropout

```python
model  = Sequential()
```

```python
model.add(Dense(units = 128, activation = 'relu', input_dim = 1))
model.add(Dropout(0.2))
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 1, activation = 'linear'))
```

```python
model.compile(optimizer = 'adam', loss = 'mse', metrics = ['accuracy'])
```

```python
dropout_history = model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 500)
```

```python
y_pred_dropouts = model.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color = 'red', label = 'Train')
plt.scatter(x_test, y_test, color = 'black', label = 'Test')
# plt.plot(x_test, y_pred, color = 'blue', label = 'Prediction')
plt.plot(x_test, y_pred_dropouts, color = 'green', label = 'Prediction with Dropout')
plt.legend()
plt.show()
```
