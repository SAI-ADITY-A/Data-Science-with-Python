This rolling average gives us the smoothening of the curve or time series data forecasting

## Simple moving average

```python
ds['Close'].plot(figsize = (20,6))
```

```python
ds['Close: 10 days window'] = ds.Close.rolling(window = 10, min_periods = 1).mean()
```

```python
ds[['Close', 'Close: 10 days window']].plot(figsize = (20,6))
plt.xlim(['2020-01-01', '2021-01-01'])
```

```python
ds['Close: 20 days window'] = ds.Close.rolling(window = 20, min_periods = 1).mean()
ds['Close: 30 days window'] = ds.Close.rolling(window = 30, min_periods = 1).mean()
```

```python
ds[['Close', 'Close: 10 days window', 'Close: 20 days window', 'Close: 30 days window']].plot(figsize = (20,6))
plt.xlim(['2020-01-01', '2021-01-01'])
```

## Cumulative moving average

```python
# Expanding function
ds['Close: cumulative moving average'] = ds.Close.expanding().mean()
```

```python
ds[['Close', 'Close: cumulative moving average']].plot(figsize = (20,6))
```

## Exponential moving average

```python
ds['EMA_0.2'] = ds['Close'].ewm(alpha = 0.2, adjust = False).mean()
```

```python
ds[['Close', 'EMA_0.2']].plot(figsize = (20,6))
```

## Exponential Weighted moving average

```python
ds['EWMA_5days'] = ds.Close.ewm(span = 5).mean()
```

```python
ds[['Close', 'EWMA_5days']].plot(figsize = (20,6))
```

```python
ds[['Close', 'EWMA_5days']].plot(figsize = (20,6))
plt.xlim(['2020-01-01', '2021-01-01'])
```

### Checking stationarity

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adf_test(ds['Close'])
```
