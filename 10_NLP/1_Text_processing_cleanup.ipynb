## 1) Basic cleanup - HTML tags from web scraping

```python
html_text = '<p>This is a paragraph with introduction to text preparation in NLP, basic cleanup.</p>'
```

```python
html_text
```

```python
import re

def scripthtml(text):
    p = re.compile(r'<.*?>')
    return p.sub('', text)
```

```python
scripthtml(html_text)
```
