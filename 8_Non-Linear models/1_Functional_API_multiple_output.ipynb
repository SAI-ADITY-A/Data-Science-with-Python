```python
from keras.models import Model
from keras.layers import Dense, Input, concatenate
```

First define the input we are giving in:


```python
x = Input(shape = (3,))
```

Next describe the hidden layers and also specify the input to the current layer

```python
hidden_1 = Dense(64, activation = 'relu')(x)
hidden_2 = Dense(32, activation = 'relu')(hidden_1)
hidden_3 = Dense(4, activation = 'relu')(hidden_2)
```

Then we branch out to solve different problems, say one is regression and other is classification

```python
output_1 = Dense(1, activation = 'linear')(hidden_3)
output_2 = Dense(1, activation = 'sigmoid')(hidden_3)
```

Now, define the model using the Model class

```python
model = Model(inputs = x, outputs = [output_1, output_2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_shapes = True)
```
